```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
projects = pd.read_csv('projectsManual.csv')
```

```python
map = folium.Map()
for i, row in projects.iterrows():
    folium.Marker(location=[row['Lat'], row['Long']], 
                       popup=str(projects.iloc[i]['Name']) + ' ' + str(projects.iloc[i]['projectID'])).add_to(map)
map
```

```python
map.save('projectMap.html')
```
